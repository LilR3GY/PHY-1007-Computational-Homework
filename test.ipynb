```python
import numpy as np

def matrice_aug(A):
    s = A.shape
    col = [[0]]*s[0]
    ligne = [0]*(s[1]+2)
    hor = np.hstack([col, A, col])
    ver = np.vstack([ligne, hor, ligne])
    return ver

matrice_aug(np.array([[1,2,3]]))
```
